```python
#### Require info
# (ok) Product ID
# Seller ID
# (ok) Product title
# (ok) Price
# (ok) URL of the product image
# (ok) URL of that product page
#### Bonus info
# (ok) Is it TikiNow (delivery within 2 hours) ?
# (ok) Is it free delivery?
# (ok) Number of reviews?
# (ok) How many stars or percentage of stars?
# (ok) Does it got "badge under price" (Rẻ hơn hoàn tiền) ?
# (ok) Discount percentage?
# Does it got "shocking price" badge ? 
# (ok) Does it allowed to be paid by installments? 
# (ok) Does it comes with free gifts? 
```

```python
def get_soup(url_head,i,url_tail) :
    '''
    Take the head and tail of tiki URL, and the page number, to combine into a whole tiki URL
    Then use selenium to get the page, and return a corresponding BeautifulSoup object
    '''
    driver = webdriver.Chrome('chromedriver',options=options)
    # we let selenium to wait for 30 seconds for all javascript script done before return the result of HTML
    driver.implicitly_wait(30) 

    tiki_url = url_head + str(i) + url_tail
    print('this is -------------------',tiki_url)
    driver.get(tiki_url)

    html_data = driver.page_source
    driver.close()
    
    # parse html into beautiful soup
    # r.text is a HTML file so use html.parser
    soup = BeautifulSoup(html_data, 'html.parser')
    
    return soup
```

```python
def extract_info_from_script(tag_object):
    '''
    This function take in a <script> tag and extract the information of the product,
    then return a dictionary of extracted info
    '''
    tag_str = str(tag_object)
    extracted_info = {'product_id':'', 'title':'', 
                      'price':'', 'image_url':'', 
                      'product_url':'', 'nStars':'',
                      'review_count':''}
    
    # find product_id
    pattern = r'.*"sku":"(\d+)".*'
    product_id1 = str(re.findall(pattern, tag_str)).strip('[')
    product_id2 = product_id1.strip(']')
    product_id = product_id2.strip("'")

    # file title
    pattern = r'"name":"(.*)","des'
    title1 = str(re.findall(pattern, tag_str)).strip('[')
    title2 = title1.strip(']')
    title = title2.strip("'")

    # find price
    pattern = r'"price":(\d+)}'
    price1 = str(re.findall(pattern, tag_str)).strip('[')
    price2 = price1.strip(']')
    price = price2.strip("'")

    # find image
    pattern = r'"image":"(.*g)","name'
    image_url1 = str(re.findall(pattern, tag_str)).strip('[')
    image_url2 = image_url1.strip(']')
    image_url = image_url2.strip("'")

    # find product_url
    pattern = r'"url":"(.*)","image"'
    product_url1 = str(re.findall(pattern, tag_str)).strip('[')
    product_url2 = product_url1.strip(']')
    product_url = product_url2.strip("'")
    
    # find number of stars
    pattern = r'"ratingValue":"([\d\.]+)"'
    nStars = re.findall(pattern, tag_str)
    if len(nStars) == 0 : 
        nStars = 0.0
    else :
        nStars = float(*nStars)
    
    # find number of review
    pattern = r'"reviewCount":"([\d\.]+)"'
    review_count = re.findall(pattern, tag_str)
    if len(review_count) == 0 : 
        review_count = 0
    else :
        review_count = int(*review_count)
    
    try:
        extracted_info['product_id'] = product_id
        extracted_info['title'] = title
        extracted_info['price'] = price
        extracted_info['image_url'] = image_url
        extracted_info['product_url'] = product_url
        extracted_info['nStars'] = nStars
        extracted_info['review_count'] = review_count
    except:
        print("We got one article error!")
    
    return extracted_info
```

```python
def extract_info_from_class_product_item(tag_object) :
    '''
    This function take in a tag of class "product-item" and extract the information of the product,
    then return a dictionary of extracted info
    '''
    extracted_info = {'product_url':'', 'title':'', 
                      'image_url':'', 'tiki_now':'',
                      'discount':'', 'badgeUnderPrice':'',
                      'freeShip':'', 'review_count':'',
                      'installment':'', 'freeGift':''}
    
    # find product_url
    product_url = 'https://tiki.vn' + tag_object['href']
    
    # find image
    image_url = tag_object.find('div', {'class':'thumbnail'}).img['src']
    
    # check TikiNow
    tikiNow1 = tag_object.find('div',{'class':'badge-service'})
    tiki_now = True if tikiNow1.div else False
    
    # find title
    title = tag_object.find('div',{'class':'name'}).span.text
    
    # find number of reviews
    review_count = 0
    reviewCount1 = tag_object.find('div',{'class':'review'})
    if reviewCount1 :
        review_count = int(reviewCount1.text[1:-1])
    
    # discount %
    discount = ''
    discount1 = tag_object.find('div',{'class':'price-discount__discount'})
    if discount1 :
        discount = discount1.text
    
    # badge under price (Re hon hoan tien)
    tagBadgeUnderPrice = tag_object.find('div',{'class':'badge-under-price'})
    badgeUnderPrice = True if tagBadgeUnderPrice.div else False
    
    # Installment (tra gop)
    tagBadgeBenefit = tag_object.find('div',{'class':'badge-benefits'})
    installment = True if tagBadgeBenefit.div else False
    
    # free gift 
    tagFreeGift = tag_object.find('div',{'class':'freegift-list'})
    freeGift = True if tagFreeGift else False
    
    # free ship
    tagBadgeTop = tag_object.find('div',{'class':'badge-top'})
    freeShip = True if tagBadgeTop.text else False
    
    try:
        extracted_info['product_url'] = product_url
        extracted_info['image_url'] = image_url
        extracted_info['tiki_now'] = tiki_now
        extracted_info['title'] = title
        extracted_info['review_count'] = review_count
        extracted_info['discount'] = discount
        extracted_info['badgeUnderPrice'] = badgeUnderPrice
        extracted_info['installment'] = installment
        extracted_info['freeGift'] = freeGift
        extracted_info['freeShip'] = freeShip
    except:
        print("We got one article error!")
    
    return extracted_info
```

```python
# request page may pha ca phe
data = {'product_id':[], 
        'title':[], 
        'price':[],
        'image_url':[], 
        'product_url':[], 
        'nStars':[],
        'review_count':[],
        'tiki_now':[],
        'discount':[],
        'badgeUnderPrice':[],
        'freeShip':[],
        'installment':[],
        'freeGift':[],
}
i = 1
head = 'https://tiki.vn/may-pha-ca-phe/c1939?page='
tail = '&src=c.1882.hamburger_menu_fly_out_banner'

empty_count = 0
while True:
    soup = get_soup(head,i,tail)
    
    
    # total number of items:  
    result_items = soup.find('div', {'class':"CategoryViewstyle__Right-bhstkd-1 jJGphj"}).div.div.div.h4.text.split()[0]
    # scripts of all item in one page
    scripts = soup.find_all('script', {'type': 'application/ld+json'})
    # product items of all item in one page
    product_items = soup.find_all('a',{'class': 'product-item'})

    nItemsBefore = len(data['product_id'])
    if int(len(data['product_id'])) >= int(result_items):
        break
    else:
        # extract data from <script> and <div class='product-item' and combine them
        for element in range(1,len(scripts)-1) : 
            script_data = extract_info_from_script(scripts[element])
            product_item_data = extract_info_from_class_product_item(product_items[element-1])
            script_data.update(product_item_data)
            
            if script_data['product_id'] in data['product_id'] :
                continue
            else : 
                for info in script_data :
                    data[info].append(script_data[info])
    
    ## additional stop condition: page with no product appear more than 100 times
    nItemsAfter = len(data['product_id'])
    if nItemsAfter == nItemsBefore :
        empty_count +=1
    else :
        empty_count = 0
    
    print(nItemsBefore, nItemsAfter,empty_count)

    if empty_count >=10 :
        break

    i += 1

```
